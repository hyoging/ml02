### replace
* replace(to_replace=비교값, value=치환값)
* NaN 값은 np.nan으로 비교

```python
import pandas as pd
```

```python
df = pd.read_csv('data/nan_test.csv')
df
```

```python
df.replace(to_replace=2, value = 200)
```

```python
df['nan_1'] == "NaN"
```

```python
import numpy as np
df['nan_1'] == np.nan
```

```python
df.replace(to_replace = np.nan , value = 3000)
```

```python
df.replace(to_replace = {"nan_2" : np.nan}, value = "nan2변경")
df
```

```python
df.replace( {"nan_1" : np.nan}, "12345", inplace = True)
df
```
